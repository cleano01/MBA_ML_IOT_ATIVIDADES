## Divisão do DataSet: treino/teste

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X = data_set.drop(0, axis=1)
```

```python
Y = data_set.iloc[ : ,  0:1]
encoder_y = LabelEncoder() 
Y = pd.DataFrame( encoder_y.fit_transform(Y))

```

```python
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

### Analisando as Corelações das Colunas 



```python
y_train.columns = ['label']
corr = pd.concat([x_train, y_train ], axis=1).corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
import seaborn as sns
import matplotlib.pyplot as pl

f, ax = pl.subplots(figsize=(30, 12))
corr = pd.concat([x_train, y_train ], axis=1).corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
```

```python
x_train.describe()
```

## Pré-processamento Dos Dados 

(variabilidade/padronizados dos dados)


```python
from sklearn.preprocessing import StandardScaler
```

```python
normalizador = StandardScaler()
x_train_norm =  pd.DataFrame(normalizador.fit_transform(x_train)) 
```

```python
x_train_norm.describe().round(3)
```

## Criando Modelo e Validando 


INICIANDO MODELO COM RANDOM FOREST

```python
#importando o modelo
from sklearn.ensemble import RandomForestClassifier
```

```python
#bibliotecas de validação
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score, confusion_matrix
```

```python
rf = RandomForestClassifier()
rf.fit(x_train_norm, y_train)

pred = rf.predict(x_train_norm)
print('accuracy', accuracy_score(y_train,pred))
print('precision', precision_score(y_train,pred,average='weighted' ))
print('recall', recall_score(y_train,pred, average='weighted'))
print('f1', f1_score(y_train,pred,average='weighted'))
confusion_matrix(y_train,pred)
```

```python
x_test_norm = normalizador.transform(x_test)
```

```python
pred = rf.predict(x_test_norm)
print('accuracy', accuracy_score(y_test,pred))
print('precision', precision_score(y_test,pred,average='weighted' ))
print('recall', recall_score(y_test,pred, average='weighted'))
print('f1', f1_score(y_test,pred,average='weighted'))
confusion_matrix(y_test,pred)
```

INICIANDO MODELO KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train_norm, y_train)
```

```python
pred = knn.predict(x_train_norm)
print('accuracy', accuracy_score(y_train,pred))
print('precision', precision_score(y_train,pred,average='weighted' ))
print('recall', recall_score(y_train,pred, average='weighted'))
print('f1', f1_score(y_train,pred,average='weighted'))
confusion_matrix(y_train,pred)
```

```python
pred = knn.predict(x_test_norm)
print('accuracy', accuracy_score(y_test,pred))
print('precision', precision_score(y_test,pred,average='weighted' ))
print('recall', recall_score(y_test,pred, average='weighted'))
print('f1', f1_score(y_test,pred,average='weighted'))
confusion_matrix(y_test,pred)
```

## Aplicando o PCA

PCA, ou principal component analysis é uma técnica para reduzir a dimensão de um conjunto de dados preservando, de certa forma, suas propriedades. 


```python
#importando a biblioteca
from sklearn.decomposition import PCA
```

```python
pca = PCA(5)
```

```python
pca.fit(x_train)
```

```python
pca.explained_variance_.round()
```

```python
pca.explained_variance_ratio_.round()
```

INICIANDO A FUNÇÃO DE DIMENSÇAO DO PCA COM DADOS DE TREINO

```python
pca_transforme= pd.DataFrame(pca.transform(x_train))
```

```python
pca_transforme.head(5)
```

```python
pca_transforme.describe()
```

INICIANDO TESTE COM PCA

```python

pca_transforme_treino= pca.transform(x_train)
pca_transforme_teste= pca.transform(x_test)


x_norm_teste_pca_treino = pd.DataFrame(normalizador.fit_transform(pca_transforme_treino))
x_norm_teste_pca_teste = pd.DataFrame(normalizador.transform(pca_transforme_teste))

```

INICANDO TESTE KNN - PCA

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python

knn.fit(x_norm_teste_pca_treino, y_train)

pred = knn.predict(x_norm_teste_pca_treino)
print('accuracy', accuracy_score(y_train,pred))
print('precision', precision_score(y_train,pred,average='micro' ))
print('recall', recall_score(y_train,pred, average='micro'))
print('f1', f1_score(y_train,pred,average='micro'))
confusion_matrix(y_train,pred)
```

```python
pred = knn.predict(x_norm_teste_pca_teste)
print('accuracy', accuracy_score(y_test,pred))
print('precision', precision_score(y_test,pred,average='micro' ))
print('recall', recall_score(y_test,pred, average='micro'))
print('f1', f1_score(y_test,pred,average='micro'))
confusion_matrix(y_test,pred)
```

INICANDO TESTE COM RANDOM FOREST - PCA

```python
rf = RandomForestClassifier()
rf.fit(x_norm_teste_pca_treino,y_train )

pred = rf.predict(x_norm_teste_pca_treino)

print('accuracy', accuracy_score(y_train,pred))
print('precision', precision_score(y_train,pred, average='weighted'))
print('recall', recall_score(y_train,pred, average='weighted'))
print('f1', f1_score(y_train,pred, average='weighted'))

confusion_matrix(y_train, pred)
```

```python
rf = RandomForestClassifier()
rf.fit(x_norm_teste_pca_teste, y_test)

pred = rf.predict(x_norm_teste_pca_teste)

print('accuracy', accuracy_score(y_test, pred))
print('precision', precision_score(y_test, pred, average='weighted'))
print('recall', recall_score(y_test, pred, average='weighted'))
print('f1', f1_score(y_test, pred, average='weighted'))

confusion_matrix(y_test, pred)
```
