```python
from sklearn.model_selection import train_test_split 
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#Importação do pacote joblib
from sklearn.externals import joblib
```

```python
#Carrega o dataset com os dados de treinamento e validacao
dados_baloes = pd.read_csv("balloons_dataset.csv", encoding="utf-8")
```

```python
dados_baloes.head(5)
```

```python
#mudados cores para números
 dados_baloes['color'].replace('YELLOW', 1, inplace=True)
    
 dados_baloes['color'].replace('PURPLE', 2, inplace=True)
```

```python
#tamanho para numero 
dados_baloes['size'].replace('SMALL', 1, inplace=True)
dados_baloes['size'].replace('LARGE', 2, inplace=True)
```

```python
# estica ou mergulha para numeros
dados_baloes['act'].replace('STRETCH', 1, inplace=True)
dados_baloes['act'].replace('DIP', 2, inplace=True)
```

```python
# adulto ou criança para numeros
dados_baloes['age'].replace('ADULT', 1, inplace=True)
dados_baloes['age'].replace('CHILD', 2, inplace=True)
```

```python
dados_baloes
```

```python
#Importação do pacote joblib para persistência do modelo
data_treino = np.array(dados_baloes[['color', 'size', 'act', 'age']])
data_classif = np.array(dados_baloes['inflated'])
x_treino, dados_val, y_treino, pols_val = train_test_split(data_treino, data_classif, test_size=0.10)
```

```python
#criando o modelo
modelo_NB = GaussianNB()
modelo_NB.fit(x_treino, y_treino)
```

```python
#Persistência do modelo no arquivo modelo_mamifero_MNB.pkl
joblib.dump(modelo_NB, 'modelo_baloes.pkl')
```
